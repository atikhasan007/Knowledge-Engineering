```
import networkx as nx
import pandas as pd


G = nx.Graph()


num_districts = int(input("Enter number of districts: "))
districts = []
for i in range(num_districts):
    name = input(f"Enter name of district {i+1}: ")
    districts.append(name)

G.add_nodes_from(districts)



# --- Take user input for roads (edges) ---
num_roads = int(input("Enter number of roads: "))
edges = []
for i in range(num_roads):
    start = input(f"Road {i+1} - starting district: ")
    end = input(f"Road {i+1} - ending district: ")
    edges.append((start, end))

G.add_edges_from(edges)1

# Show results
print("\nDistricts (Nodes):", list(G.nodes))
print("Roads (Edges):", list(G.edges))

```

```
# 1. Adjacency List
print("Adjacency List:")
for node, neighbors in G.adjacency():
    print(f"{node}: {list(neighbors)}")

print("\n")
```

```

# 2. Edge List (as a pandas DataFrame for easier analysis)
print("Edge List (DataFrame):")
edge_list_df = nx.to_pandas_edgelist(G)
display(edge_list_df)
```

```
# 3. Adjacency Matrix
print("Adjacency Matrix:")
adj_matrix = nx.adjacency_matrix(G)
print(adj_matrix.todense())

# Alternatively, as a pandas DataFrame for better readability with node labels:
adj_df = nx.to_pandas_adjacency(G)
print("\nAdjacency Matrix (DataFrame):")
display(adj_df)
```

```
# Function to add a road between two districts
def add_road(graph, district1, district2):
    """Adds an edge between two nodes in the graph."""
    if district1 in graph.nodes and district2 in graph.nodes:
        if not graph.has_edge(district1, district2):
            graph.add_edge(district1, district2)
            print(f"Road added between {district1} and {district2}.")
        else:
            print(f"Road already exists between {district1} and {district2}.")
    else:
        print("One or both districts not found in the network.")
```

```

# Function to remove a road between two districts
def remove_road(graph, district1, district2):
    """Removes an edge between two nodes in the graph."""
    if graph.has_edge(district1, district2):
        graph.remove_edge(district1, district2)
        print(f"Road removed between {district1} and {district2}.")
    else:
        print(f"No road found between {district1} and {district2}.")
```

```



# Add a road between 'Hospital' and 'Police Station'
print("Attempting to add road between Hospital and Police Station:")
add_road(G, 'Hospital', 'Police Station')

# Display updated Adjacency List
print("\nUpdated Adjacency List:")
for node, neighbors in G.adjacency():
    print(f"{node}: {list(neighbors)}")

# Display updated Adjacency Matrix (DataFrame)
print("\nUpdated Adjacency Matrix (DataFrame) after adding road:")
updated_adj_df_after_add = nx.to_pandas_adjacency(G)
display(updated_adj_df_after_add)


# Remove the road between 'Suburb A' and 'Suburb B'
print("\nAttempting to remove road between Suburb A and Suburb B:")
remove_road(G, 'Suburb A', 'Suburb B')

# Display updated Adjacency List
print("\nUpdated Adjacency List:")
for node, neighbors in G.adjacency():
    print(f"{node}: {list(neighbors)}")

# Display updated Adjacency Matrix (DataFrame)
print("\nUpdated Adjacency Matrix (DataFrame) after removing road:")
updated_adj_df_after_remove = nx.to_pandas_adjacency(G)
display(updated_adj_df_after_remove)
```

```
# 4. Connectivity Check
print("Connectivity Check:")

if nx.is_connected(G):
    print("The city network is fully connected.")
else:
    print("The city network is NOT fully connected.")
    # You can also find the connected components if it's not connected
    print("Connected components:", list(nx.connected_components(G)))
```

```
# 5. List reachable districts
def list_reachable_districts(graph, start_district):
    """Lists all reachable districts from a starting district."""
    if start_district not in graph.nodes:
        print(f"Error: '{start_district}' not found in the network.")
        return

    reachable_nodes = set(nx.nodes(graph)) # Initialize with all nodes

    # For an undirected graph, all nodes in the connected component of the start_district are reachable.
    if nx.is_connected(graph):
        print(f"All districts are reachable from '{start_district}' (network is fully connected).")
        print("Reachable districts:", list(reachable_nodes))
    else:
        # Find the connected component containing the start node
        component = None
        for comp in nx.connected_components(graph):
            if start_district in comp:
                component = comp
                break
        if component:
            print(f"Reachable districts from '{start_district}':")
            print(list(component))
        else:
             # This case should not be reached if start_district is in graph.nodes
             print(f"Could not find the connected component for '{start_district}'.")


# Example usage:
start_district = input("Enter the starting district to list reachable districts: ")
list_reachable_districts(G, start_district)
```

```
4# 6. Shortest Route for Rescue (minimum number of nodes)
def find_shortest_route_nodes(graph, start_district, end_district):
    """Finds the shortest path (minimum number of nodes) between two districts."""
    if start_district not in graph.nodes:
        print(f"Error: Start district '{start_district}' not found in the network.")
        return
    if end_district not in graph.nodes:
        print(f"Error: End district '{end_district}' not found in the network.")
        return

    if not nx.has_path(graph, start_district, end_district):
        print(f"No path exists between '{start_district}' and '{end_district}'.")
        return

    # Find the shortest path in terms of number of nodes (unweighted)
    shortest_path = nx.shortest_path(graph, source=start_district, target=end_district)

    print(f"Shortest path from '{start_district}' to '{end_district}':")
    print(" -> ".join(shortest_path))
    print(f"Number of nodes in the shortest path: {len(shortest_path)}")
    print(f"Number of roads (edges) in the shortest path: {len(shortest_path) - 1}")


# Example usage:
start_district = input("Enter the starting district for the shortest route: ")
end_district = input("Enter the ending district for the shortest route: ")

find_shortest_route_nodes(G, start_district, end_district)
```

```
# 7. Find all paths between two districts
def find_all_paths(graph, start_district, end_district):
    """Finds and displays all possible paths between two districts."""
    if start_district not in graph.nodes:
        print(f"Error: Start district '{start_district}' not found in the network.")
        return
    if end_district not in graph.nodes:
        print(f"Error: End district '{end_district}' not found in the network.")
        return

    if not nx.has_path(graph, start_district, end_district):
        print(f"No path exists between '{start_district}' and '{end_district}'.")
        return

    print(f"All simple paths from '{start_district}' to '{end_district}':")
    # nx.all_simple_paths returns an iterator of lists of nodes
    for path in nx.all_simple_paths(graph, source=start_district, target=end_district):
        print(" -> ".join(path))

# Example usage:
start_district_all = input("Enter the starting district to find all paths: ")
end_district_all = input("Enter the ending district to find all paths: ")

find_all_paths(G, start_district_all, end_district_all)
```

```
# 8. Find Critical Districts (Articulation Points)
print("Critical Districts (Articulation Points):")

# An articulation point or cut vertex is a node whose removal increases the number of connected components.
# These are crucial in a network as their failure can lead to disconnection.
articulation_points = list(nx.articulation_points(G))

if articulation_points:
    print("The following districts are critical (articulation points):")
    print(articulation_points)
else:
    print("There are no articulation points in the network. Removing any single district will not disconnect the network.")
```

```
# 9. Degree Analysis
print("Degree Analysis:")

# Get the degree of each node
degrees = dict(G.degree())
print("Degree of each district:")
print(degrees)

# Find the district(s) with the highest degree
max_degree = max(degrees.values())
districts_highest_degree = [district for district, degree in degrees.items() if degree == max_degree]
print(f"\nDistrict(s) with the highest degree ({max_degree}): {districts_highest_degree}")

# Find the district(s) with the lowest degree
min_degree = min(degrees.values())
districts_lowest_degree = [district for district, degree in degrees.items() if degree == min_degree]
print(f"District(s) with the lowest degree ({min_degree}): {districts_lowest_degree}")
```
