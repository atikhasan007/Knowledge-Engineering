```
import networkx as nx
import matplotlib.pyplot as plt
#Create an undirected Graph
G=nx.Graph()

#Add Nodes
G.add_nodes_from([1,2,3,4,5,6])

#Add Edges
G.add_edges_from([(1,1), (1,2), (1,3), (2,3),
                  (3,4), (4,5),(1,5),(3,5)])
#Draw the grpah
nx.draw(G,with_labels=True, font_weight='bold')
plt.show()
```

```
#Basic Graph Information

print("Nodes: ",G.nodes())
print("Edges: ",G.edges())
print("Number of Nodes: ",G.number_of_nodes())
print("Number of Edges: ", G.number_of_edges())
```

```
#check Graph Connectivity

print("Is Graph Connected?: ",nx.is_connected(G))
```

```
#shortest Path between two nodes

shortest_path=nx.shortest_path(G, source=1, target=4)
print("Shortest Path from Node 1 to 4:",shortest_path)
```

```
#Degree of Each Node

degree_seq= G.degree()
print("Degree of each Node: ",degree_seq)
```

```
#Clustering Co-efficent of Each Node

cluster_coefficent=nx.clustering(G)
print("Clustering Coefficient of each node: ", cluster_coefficent)
```

```
#draw a subgrpah from main Graph

subgraph_nodes=[1,2,3]
subgraph= G.subgraph(subgraph_nodes)
nx.draw(subgraph, with_labels=True, font_weight='bold')
```

```
nx.is_connected(subgraph)
```

```
#Duplicate Graph
G_copy=G.copy()
nx.draw(G_copy, with_labels=True, font_weight='bold')
```

```
#Remove a node and its associated Edges

G_copy.remove_node(4)
nx.draw(G_copy, with_labels=True, font_weight='bold')
plt.show()
```

```
#Remove a specific Edge
G_copy.remove_edge(1,5)
nx.draw(G_copy, with_labels=True, font_weight='bold')
plt.show()
```

```
#Find Connected Components in Graph

components=list(nx.connected_components(G_copy))
print("Connected Components:",components)

for i, component in enumerate(components):
  subgraph=G_copy.subgraph(component)
  plt.figure()
  nx.draw(subgraph, with_labels=True, font_weight='bold')
  plt.title(f"Connected Component{i+1}")
  plt.show()
```

```
#Density Calculation

density=nx.density(G)
print("Graph Density: ", density)
```

```
#visualize the degree distribution of Graph

degree_sequence=[G.degree(node) for node in G. nodes()]
plt.hist(degree_sequence,bins='auto')
plt.title("Degree Distribution")
plt.xlabel("Deegree")
plt.ylabel("Frequency")
plt.show()
```

```
#regions

regions=list(nx.connected_components(G_copy))
G_copy.remove_node(6)
regions=list(nx.connected_components(G_copy))
print("Regions(Connected components):",regions)
```

```
#Length of the graph

graph_length=nx.algorithms.distance_measures.diameter(G_copy)
print("Length of the Graph (Diameter): ", graph_length)
```

```
#ecentricity of each node
ecentricity=nx.eccentricity(G_copy)
print("Ecentricity of each Node: ", ecentricity)
```

```
#diameter
diameter=nx.diameter(G_copy)
print("Diameter of the Graph: ",diameter)
```

```
#radius
radius=nx.radius(G_copy)
print("Radius of the Graph: ",radius)
```

```
#Center of the graph
center=nx.center(G_copy)
print("Center of the Graph: ",center)
```

```
#Peripheral Nodes of the Graph
peripheral= nx.periphery(G_copy)
print("Peripheral Nodes of the graph: ", peripheral)
```
